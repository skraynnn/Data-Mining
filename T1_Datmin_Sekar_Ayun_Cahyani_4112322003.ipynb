**MELAKUKAN PEMBERSIHAN DATA**

```python
movdf.loc[:, 'color'] = movdf['color'].str.strip().str.capitalize()
movdf.loc[:, 'country'] = movdf['country'].str.strip().str.upper()
movdf = movdf[(movdf['duration']>=0)&
              (movdf['gross']>=0)&
              (movdf['title_year']>=0)&
              (movdf['budget']>=0)&
              (movdf['imdb_score']>=0)&
              (movdf['movie_facebook_likes']>=0)]
movdf = movdf.replace('N/A', pd.NA)
movdf.dropna(inplace=True)
movdf
```

```python
print(movdf.isnull().sum())
```

Dapat dilihat baris yang memiliki nilai NaN di kolom penting seperti gross dan budget sudah terhapus, nilai yang tidak konsisten atau kesalahan penulisan di kolom, seperti perbedaan antara "Color" dan "color" juga sudah diatasi, dan nilai-nilai yang tidak standar, seperti nilai negatif atau "N/A" sudah diubah dan dihapus. Hal itu dapat dilihat dengan kode diatas ini


**MELAKUKAN TRANSFORMASI DATA**

Melakukan pengecekan data

```python
movdf.dtypes
```

Mengubah Tipe datanya

```python
movdf['gross'] = movdf['gross'].astype('int')
movdf['budget'] = movdf['budget'].astype('int')
movdf.dtypes
```

Selanjutnya melakukan normalisasi text

```python
movdf['actors'] = movdf['actors'].str.replace(',', ', ', regex=False)
movdf['genres'] = movdf['genres'].str.replace('|', ', ', regex=False)
movdf
```

**MELAKUKAN PENYIMPANAN DATA**

```python
movdf.to_csv('movie_sample_dataset_cleaned.csv', index=False)
```
